```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=0)

# Create x values
x = np.linspace(-5, 5, 100)

# Compute activation function values
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
y_softmax = softmax(x)

# Plot activation functions
plt.figure(figsize=(16, 6))

plt.subplot(1, 4, 1)
plt.plot(x, y_sigmoid, label='Sigmoid', color='b')
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()

plt.subplot(1, 4, 2)
plt.plot(x, y_relu, label='ReLU', color='g')
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.grid(True)
plt.legend()

plt.subplot(1, 4, 3)
plt.plot(x, y_tanh, label='Tanh', color='r')
plt.title('Tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()

plt.subplot(1, 4, 4)
plt.plot(x, y_softmax, label='Softmax', color='m')
plt.title('Softmax')
plt.xlabel('x')
plt.ylabel('softmax(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
